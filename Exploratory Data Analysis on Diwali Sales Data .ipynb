## Exploratory Data Analysis

```python
df.columns
```

```python
ax = sns.countplot(x='Gender', data = df)         #Great but no labels
```

```python
ax = sns.countplot(x='Gender', data = df)   #Now add labels

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
print(df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False))
#Groups amount by gender
```

```python
SalesByGender = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x='Gender', y ='Amount', data = SalesByGender)
```

```python
ax = sns.countplot(x='Age Group', data = df)   #without hue

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x='Age Group', data = df, hue = 'Gender')   #with hue, adds legend into your graph

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)

#We can even add the age group column if it does not already exists
```

```python
SalesByAge = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
                                                                       
sns.barplot(x='Age Group', y ='Amount', data = SalesByAge)
```

```python
SalesByAge = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
                                                                       
sns.barplot(x='Age Group', y ='Amount', data = SalesByAge)
```

```python
#Top 8 order count cities

OrderCountByState = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(8)

sns.set(rc ={'figure.figsize':(18,7)})
sns.barplot(x='State', y='Orders', data = OrderCountByState)
```

```python
#Top 8 order count cities

OrderCountByState = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(8)

sns.set(rc ={'figure.figsize':(18,7)})
sns.barplot(x='State', y='Orders', data = OrderCountByState)
```

```python
#Top 8 sales states

RevenueByState = df.groupby(['State'], as_index = False )['Amount'].sum().sort_values(by='Amount', ascending = False).head(8)

sns.barplot(x='State', y ='Amount', data = RevenueByState)
```

```python
ax = sns.countplot(x='Marital_Status', data = df, hue ='Gender')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Count based on occupation
ax = sns.countplot(x ='Occupation', data =df)

sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize': (20,5)})
sns.barplot(data = sales_state, x = 'Occupation', y = 'Amount')
```

```python
sns.set(rc = {'figure.figsize': (20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

top_10_states = sales_state.head(10)

sns.set(rc={'figure.figsize': (20, 5)})

sns.barplot(data=top_10_states, x='Product_Category', y='Amount')
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

top_10_states = sales_state.head(10)

sns.set(rc={'figure.figsize': (20, 5)})

sns.barplot(data=top_10_states, x='Product_ID', y='Orders')
```
